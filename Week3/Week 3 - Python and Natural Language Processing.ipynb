#### 1.4.2 Matrix
Matrices are two-dimensional arrays of the same data type, and are commonly used to store structured data such as numbers, word frequencies, or pixel values in images. In Python, matrices can be categorized into two main types:

- Dense Matrix: A matrix where most of the values are non-zero.

- Sparse Matrix: A matrix where many of the values are zero. Storing such matrices as dense structures can waste memory, so special formats are used to save only the non-zero values and their positions.

Matrices in Python are commonly created and manipulated using **NumPy** and **scipy**, two powerful packages for numerical computing.  Did you notice that we used the command ```import numpy as np``` in the code? This is the very first time for us to import a package. Packages are an essential building block in programming. Below we show the creation of a dense matrix and its corresponding sparse matrix. 

````python
import numpy as np

# Create a dense 3x6 matrix filled with zeros
dense_matrix = np.zeros((3, 6))

# Fill in the non-zero values
dense_matrix[0, 2] = 0.7959605415681652
dense_matrix[0, 4] = 0.6053485081062916

dense_matrix[1, 0] = 0.7959605415681654
dense_matrix[1, 4] = 0.6053485081062917

dense_matrix[2, 1] = 0.5773502691896257
dense_matrix[2, 3] = 0.5773502691896257
dense_matrix[2, 5] = 0.5773502691896257

print("Dense matrix:\n", dense_matrix)
````

### 2.1 A Simple Example

Before analyzing real Twitter data, let's start with a simple example to understand the process.

Here's a small collection (called a "corpus") of three short documents:

- **Document 1:** "It is going to rain today."
- **Document 2:** "Today I am not going outside."
- **Document 3:** "I am going to watch the season premiere."

**Our Task:** Convert these sentences into numbers that capture their meaning.

````python
# Let's store our three documents in a list
Document1 = "It is going to rain today."
Document2 = "Today I am not going outside."
Document3 = "I am going to watch the season premiere."

# Combine them into a corpus (collection of documents)
Doc = [Document1, Document2, Document3]

print("Our corpus contains", len(Doc), "documents:")
for doc in Doc:
    print(doc)
````


From this example, we’ll learn how to convert raw text into numerical features — or what we might call columns of numbers. This process is often referred to as vectorization.

Once we represent text as vectors, we unlock the ability to perform various types of analysis, including:
- Summarization
- Clustering
- Topic modeling
- Information retrieval (e.g., finding similar texts)
- Predictive modeling

The core idea in Natural Language Processing (NLP) is transforming unstructured text into structured numerical form. While there are many ways to do this, we’ll focus on one of the most widely used and interpretable methods: TF-IDF (Term Frequency–Inverse Document Frequency).

TF-IDF is useful in many NLP applications. For example:
- Search engines use it to rank the relevance of a document to a search query.
- It’s also used in text classification, summarization, and topic modeling.

After learning TF-IDF, we’ll apply it in a downstream task — topic modeling — to uncover hidden themes across the documents.

While we won’t cover every vectorization technique or downstream task, this example will give you a strong foundation for understanding how an NLP pipeline works.


### 2.2 Vectorization: Term Frequency (TF) — Inverse Document Frequency (IDF)
A corpus can be defined as a collection of documents. In our example, each sentence is a document, and they collectively form a corpus.  

To vectorize text data, we use a TF-IDF method. 
- We first tokenize the text, and then assign an importance score for every term. 
- The importance score of a term is high when it occurs a lot in a given document and rarely in others. 
- In short, commonality within a document measured by TF is balanced by rarity between documents measured by IDF. The resulting TF-IDF score reflects the importance of a term for a document in the corpus.
 

````python
# Step 1: Import the TF-IDF tool from sklearn (a popular machine learning library)
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Step 2: Create a vectorizer
vectorizer = TfidfVectorizer()

# Step 3: First, let's see how the text is broken into words (tokenization)
analyze = vectorizer.build_analyzer()
print("Document 1 tokens:", analyze(Document1))
print("Document 2 tokens:", analyze(Document2))
print("Document 3 tokens:", analyze(Document3))

# Step 4: Transform our documents into TF-IDF vectors
X = vectorizer.fit_transform(Doc)

print("\nTF-IDF Matrix (sparse format):")
print(X)

# Step 5: Convert to a readable DataFrame

df= pd.DataFrame.sparse.from_spmatrix(X,columns=vectorizer.get_feature_names_out())

#df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())
print("\nTF-IDF Matrix (readable format):")
df
````


We tokenize and generate a vocab of the document. For each document, we can find the TF= (Number of repetitions of word in a document) / (# of words in a document). We can further find the IDF=Log[(Number of documents) / (Number of documents containing the word)]

| words      | Doc1 | Doc2| Doc3|IDF Value|
| ----------- | ----------- |----------- |----------- |----------- |
| going      | 0.16     |0.16|0.12|0|
| to   | 0.16       |0|0.12|0.41|
|today|0.16|0.16|0|0.41|
|i|0|0.16|0.12|0.41|
|am|0|0.16|0.12|0.41|
|it|0.16|0|0|1.09|
|is |0.16|0|0|1.09|
|rain|0.16|0|0|1.09|

We then construct a document-term matrix using the TF-IDF scores:

| Docs      | going |to|today|i|am|it|is|rain|
| ------ |------ |------ |------ |------ |------ |------ |------ |------ |
| Doc1      | 0  |0.07|0.07|0|0|0.17|0.17|0.17|0.17|
| Doc2   | 0  |0|0.07|0.07|0.07|0|0|0|
|Doc3|0|0.05|0|0.05|0.05|0|0|0|

It is easy to see that 'it', 'is', and 'rain' are important for Doc 1 but not Doc 2 or Doc 3. Each row of the document-term matrix can be thought of as a numeric representation of the documents, which we often term vectors. These numeric representations help you to find similarities between documents. 
 
> You might have noticed that stop words such as “to” and “is” are included above. These are usually filtered out in real-world NLP tasks because they don’t carry much meaning.

To perform vectorization in Python, we use the <code>TfidfVectorizer</code> from the <code>sklearn</code> package.

The steps are:
- Create the vectorizer.
- Fit it on your corpus.
- Transform your corpus into vectors.

The function **TfidfVectorizer** takes two parameters. 
- max_df is used for removing terms that appear too frequently, also known as "corpus-specific stop words". For example:
    - max_df = 0.50 means "ignore terms that appear in more than 50% of the documents".
    - max_df = 25 means "ignore terms that appear in more than 25 documents".
    - The default max_df is 1.0, which means "ignore terms that appear in more than 100% of the documents". Thus, the default setting does not ignore any terms.
- min_df is used for removing terms that appear too infrequently. For example:
    - min_df = 0.01 means "ignore terms that appear in less than 1% of the documents".
    - min_df = 5 means "ignore terms that appear in less than 5 documents".
    - The default min_df is 1, which means "ignore terms that appear in less than 1 document". Thus, the default setting does not ignore any terms.



````python
# Now let's remove common words (stop words) to focus on meaningful terms

# Step 1: Create a vectorizer with stop words removed and adjusted parameters
tfidf_vectorizer = TfidfVectorizer(
    max_df=0.95,  # Ignore words appearing in more than 95% of documents
    min_df=0.1,   # Ignore words appearing in less than 10% of documents
    stop_words='english'  # Remove common English words like "the", "is", "and"
)

# Step 2: Transform the documents
tfidf = tfidf_vectorizer.fit_transform(Doc)

# Step 3: View the result
tfidf_df = pd.DataFrame.sparse.from_spmatrix(tfidf, columns=tfidf_vectorizer.get_feature_names_out())
print("TF-IDF after removing stop words:")
tfidf_df
````

Now, after removing stop words, the resulting matrix looks like this (we'll call it M):

|     | outside   | premiere | rain     | season   | today    | watch    |


### 2.3 Topic Modeling with Non-negative Matrix Factorization (NMF)

####  The Big Picture: Why Topic Modeling?

Imagine you own an e-commerce store with thousands of products. Instead of analyzing each product individually, you might want to **group them into categories** like "Electronics," "Clothing," "Home & Garden," etc. This makes it much easier to understand your inventory and customer preferences.

**Topic modeling does the same thing for text!** Instead of dealing with hundreds or thousands of individual words, we group them into a smaller number of meaningful "topics." Each topic is a cluster of related words.

**Business Application:** 
- Analyze customer reviews to find common themes (quality issues, shipping complaints, product features)
- Understand what your social media audience talks about most
- Categorize support tickets automatically
- Identify trending topics in your industry

---

####  How NMF Works: The Mathematics

TF-IDF vectors are great, but they're **high-dimensional** — when we have hundreds or thousands of unique words, it's hard to make sense of them all.

**Non-negative Matrix Factorization (NMF)** breaks down our large document-term matrix into two smaller, more interpretable matrices:

$$
M = W \times H
$$

Where:
- **M**: Original document-term matrix (m documents × n terms)
- **W**: Document-topic matrix (m documents × k topics) — *How much does each document belong to each topic?*
- **H**: Topic-term matrix (k topics × n terms) — *Which words define each topic?*
- **k**: Number of topics (we choose this!)

**Key constraint:** All values must be **non-negative** (≥ 0), which makes the results more interpretable.

---

#### Interpreting the Results

After running NMF, you get two matrices:

1. **W Matrix (Document-Topic)**: Shows the "strength" of each topic in each document
   - High values = that document strongly belongs to that topic
   - Low values = that document barely belongs to that topic

2. **H Matrix (Topic-Term)**: Shows which words define each topic
   - High values = that word is very important for that topic
   - Low values = that word doesn't really belong to that topic

---

####  NMF in Python

The `NMF` function from scikit-learn takes two key parameters:
- **n_components**: The number of topics you want to extract
- **random_state**: Controls the random number generator (set it to get reproducible results)


````python
from sklearn.decomposition import NMF

# Step 1: Create the NMF model with 2 topics
nmf_model = NMF(n_components=2, random_state=0)

# Step 2: Fit the model and get the W matrix (document-topic matrix)
W = nmf_model.fit_transform(tfidf)

# Step 3: Display what each topic is about by finding the top words
feature_names = tfidf_vectorizer.get_feature_names_out()
topic_names = []

print("Discovered Topics:\n")
for topic_index, topic in enumerate(nmf_model.components_):
    # Get the top 3 words for this topic
    top_words = pd.Series(topic, index=feature_names).nlargest(3).index.tolist()
    top_words_string = " ".join(top_words)
    print("Topic #"+str(topic_index)+ ": "+ top_words_string)
    topic_names.append(top_words_string)

# Step 4: Create a readable DataFrame showing document-topic strengths
print("W Matrix - How much does each document belong to each topic?")
topic_df = pd.DataFrame(W, columns=topic_names)
display(topic_df)

# Step 5: Show the H matrix - which words define each topic
print(" H Matrix - Which words define each topic?")
h_matrix_df = pd.DataFrame(nmf_model.components_, columns=feature_names)
display(h_matrix_df)
````

### 3.5 Linear Regression: Which Topics Drive Engagement?

---

#### The Business Question

We now have:
- **Topics** identified from tweets (via NMF)
- **Engagement data** (likes, views, etc.)

**The key question:** Which topics lead to more engagement?

If we know that "Topic 3: AI writing tips" gets more likes than "Topic 5: product updates," we should create more content about AI writing tips!

---

#### What is Linear Regression?

**Linear regression** is one of the most fundamental techniques in data analysis. It helps us understand and quantify relationships between variables.

**In simple terms:** It draws a line (or hyperplane in multiple dimensions) that best predicts an outcome based on input features.

**For our analysis:**
- **Input (X):** Topic weights for each tweet (from NMF)
- **Output (Y):** Engagement metric (like count, views, etc.)
- **Goal:** Understand which topics have the strongest positive or negative effect on engagement

---

#### The Mathematical Model

Linear regression assumes a relationship of this form:

$$
\text{Engagement} = \beta_0 + \beta_1 \cdot \text{Topic}_1 + \beta_2 \cdot \text{Topic}_2 + \dots + \beta_k \cdot \text{Topic}_k
$$

Where:
- **$\beta_0$** (intercept): The baseline engagement when all topic weights are zero
- **$\beta_1, \beta_2, \dots, \beta_k$** (coefficients): The effect of each topic on engagement
  - **Positive coefficient** ($\beta_i > 0$): This topic is associated with **MORE engagement** 
  - **Negative coefficient** ($\beta_i < 0$): This topic is associated with **LESS engagement** 
  - **Coefficient near zero**: This topic has little impact on engagement

---

#### Interpretation

After running the regression, you'll get coefficients for each topic. Here's how to interpret them:

| Coefficient | Interpretation | Action |
|-------------|----------------|--------|
| **+0.50** | Strong positive effect | **Create more content about this topic!** |
| **+0.10** | Weak positive effect | This topic helps a little |
| **~0.00** | No effect | This topic doesn't impact engagement |
| **-0.20** | Negative effect | Consider reducing content about this topic |

**Example:**
- Topic 2 (AI writing tips): coef = +0.45 → **High engagement! Prioritize this content**
- Topic 5 (product updates): coef = -0.12 → **Low engagement; maybe reframe these posts**

---

#### Running Linear Regression in Python

We'll use the `pingouin` package, which provides easy-to-read regression output including:
- **Coefficients:** How much each topic affects engagement
- **P-values:** Whether the effect is statistically significant (p < 0.05 means significant)


````python
import pingouin as pg

# Step 1: Create a combined dataframe with topics (X) and engagement (Y)
df_model = topic_df.copy()

# Step 2: Calculate engagement ratio (likes per view)
# This normalizes for the reach of the tweet
df_model['engagement_ratio'] = df['public_metrics.like_count'] / df['public_metrics.impression_count']

# Step 3: Run linear regression
# X = all topic columns, Y = engagement_ratio
result = pg.linear_regression(
    df_model.drop(columns='engagement_ratio'),  # Input features (topics)
    df_model['engagement_ratio']                 # Output target (engagement)
)

# Step 4: Display the key results
# - 'names': Variable name (topic)
# - 'coef': Coefficient (positive = more engagement, negative = less)
# - 'pval': P-value (< 0.05 means statistically significant)
result_display = result[['names', 'coef', 'pval']].round(3)

print(" Regression Results: Which Topics Drive Engagement?\n")
print(result_display)

print("\n How to read this:")
print("- Positive 'coef' = this topic increases engagement")
print("- Negative 'coef' = this topic decreases engagement")
print("- 'pval' < 0.05 = the effect is statistically significant")
````
